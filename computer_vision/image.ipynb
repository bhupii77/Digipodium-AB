```python
cat = cv2.imread("cat.jpg")
```

```python
plt.imshow(cat)
plt.axis('off')
plt.show()
```

```python
cat_color_converted = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
plt.imshow(cat_color_converted)
plt.axis('off')
```

```python
cat_color_grey = cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)
plt.imshow(cat_color_grey, cmap='gray')
plt.axis('off')
```

```python
plt.imshow(np.hstack([cat, cat_color_converted]))
plt.axis('off')
```
